We load it up and have a look at it

```python
import pandas

education = pandas.read_csv("../data/world_bank/education.csv")

education
```

```python
countries = education["Country Name"].drop_duplicates()
countries
```

We have the choice now to either adjust our set of arab world country names or to adjust the original table.  I think it is better to take the World Bank names, so we adjust our list of arab world country names

Work out which list methods will help you make these changes and make them.

```python
# solution
arab_countries.remove("Egypt")
arab_countries.append("Egypt, Arab Rep.")
arab_countries.remove("Yemen")
arab_countries.append("Yemen, Rep.")
arab_countries.remove("Syria")
arab_countries.append("Syrian Arab Republic")
arab_countries.remove("Egypt")
print(arab_countries)
```

We can now filter our data to include only the arab world data.  At this time we will also filter to just the metric we are interested in.  You may have noticed (if you didn't, go back and look) that in this dataset there is one row for each country and each metric.  We have a big list of countries we are interested in, but just one metric.

We will use two masks for this, and join them with `&` so _both_ are applied.  Note, I have worked out the metric name for you - it is "Literacy rate, adult female (% of females ages 15 and above)"

```python
mask =  (education["Country Name"].isin(arab_countries)) & (education["Series Name"] == "Literacy rate, adult female (% of females ages 15 and above)")
data = education[mask]
data
```

We are aiming for a scatter plot, and those only work when the x values are in one column and the y values are in another column, so we need to adjust our table.

Were we working with the original table, we would be out of luck, but we have made an important adjustment to the data - there is now only one row for each country code.  If we remove uneeded columns, we get the exact transpose of what we need, if only the Country Name were the index!

```python
cleaned = data.drop(["Series Name",	"Series Code", "Country Code"], axis=1)
cleaned
```

We can achieve the exact transpose of what we need by adjusting the index.

```python
cleaned = cleaned.set_index("Country Name", drop=True)
cleaned
```

Now we only need to transpose the data to get what we need.

```python
ready = cleaned.transpose()
ready
```

we have two problems left:
  * the index is not a number, it is a string containing a number, but we need an actual number.
  * the x-coordinate is an index but we want it to be a column so we can use it in the scatter plot

## Dealing with both problems at once

We need to use a string method on each value in the index and to then set what comes back to a new column which we will call "year"

```python
ready["year"] = ready.index.map(lambda s: int(s[:4]))
ready
```

We are ready to create our scatter plot.  If we wanted to plot just one our job would be easy

```python
ready.plot(kind="scatter", x = "year", y="Arab World")
```

Something has gone wrong!  In fact, the column we are trying to plot are not numbers.  The "Arab World" column is a string, including a heap of ".." values.  The "year" column has come over as an integer though.

```python
ready.info()
```

The solution will come from `astype` however we can't apply this while there are ".." and "" values in the data.  We need to adjust those to being just an empty string.


```python
ready = ready.replace("..","nan")
ready = ready.replace("", "nan")
ready = ready.astype(float)
ready.plot(kind="scatter", x = "year", y="Arab World")
```

We can now plot every column that we have

```python
ax = pandas.Series([], dtype=float).plot()
for c in arab_countries:
  ready.plot(kind="scatter", x = "year", y=c, ax=ax)
```
